```python
#1번
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
import pandas as pd 
import time
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python

dogShieldList=[]
driver=wb.Chrome()
# 1~26페이지 이동
for page in range(1,27):
    driver.get('https://www.animal.go.kr/front/awtis/institution/institutionList.do?totalCount=251&pageSize=10&menuNo=1000000059&&page='+str(page))
    for i in range(1,11):
        #2차원배열로 만들기
        temp=[]
        for j in range(1,5):
            #유기동물보호소 정보 크롤링
            ugi=driver.find_element_by_css_selector('tbody>tr:nth-child('+str(i)+')>td:nth-child('+str(j)+')').text
            temp.append(ugi)
        dogShieldList.append(temp)
```

```python
dogShieldList
```

```python
#데이터프레임으로 만들기
dogShield=pd.DataFrame(dogShieldList)
```

```python
dogShield
```

```python
#컬럼명 바꾸기
dogShield.rename(columns={
    dogShield.columns[0]:'관할구역',
    dogShield.columns[1]:'보호센터명',
    dogShield.columns[2]:'전화번호',
    dogShield.columns[3]:'보호센터주소'
},inplace=True)
dogShield.head()
```

```python
dogShield.to_csv('유기동물보호소리스트.csv',encoding='euc-kr')
```
