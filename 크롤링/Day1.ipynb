# Request 사용법
- 페이지 정보를 요청할때 사용하는 라이브러리

```python
import requests as req
```

```python
# 네이버 페이지를 요청
# 요청을 도와주는 라이브러리 req
res = req.get("https://www.naver.com/")

# response[200] >> 페이지를 잘 받아왔습니다!
#변수에 담기전에 데이터가 잘 넘어왔는지 확인하고 변수 담기
```

```python
# 요청한 페이지의 정보를 보기!
res.text
```

# Melon 홈페이지 정보 가지고오기

```python
# response[406] >> 응답을 할수가 없습니다!
req.get("https://www.melon.com/")
```

```python
res = req.get("https://www.melon.com/",headers = h)
```

```python
res.text
```

# Beautifulsoup 사용하기
-- 가져온 데이터에서 내가 필요한 정보들만 추출해주는 라이브러리

```python
from bs4 import BeautifulSoup as bs 
```

```python
#bs(어떤 데이터를 처리할건지, 어떻게 처리=lxml)
soup = bs(res.text,'lxml')
```

```python
# 내가 가지고온 데이터에서 A태그중에 클래스명 nav만 가지고 와줘
result=soup.find_all('a',class_="nav") #result값이 list로 넘어옴 class값으로 요청해서 데이터를 받아올때는 모든 class값이 넘어오기 때문에 list로 넘어옴
```

```python
# 가지고온 데이터에서 순수하게 글자만 추출할때는 .text 기능 활용!
result[2].text
```

```python
result[0].text 
```

# 리스트의 전체 데이터를 조회

```python
for i in result:
    print(i.text)
```

# 네이버 뉴스 타이틀만 가져오기
