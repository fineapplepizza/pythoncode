#### 강아지 정보 크롤링

```python
#페이지 6~14 강아지 정보 크롤링
cntpage = 6
dog_list=[]
cntindex = 0
driver=wb.Chrome()
for p in range(0,9):
    driver.get("https://terms.naver.com/list.naver?cid=42883&categoryId=44358&so=date.dsc&viewType=&categoryType=&page="+str(cntpage))
    
    for i in range(1,16):
        btnDog=driver.find_element_by_css_selector("ul.content_list>li:nth-child("+str(i)+")>div.info_area>div.subject>strong.title>a:first-child")
        btnDog.click()
        time.sleep(1)
        soup=bs(driver.page_source,'lxml')

    #새로운 강아지 정보를 담아주기 위한 초기화
        temp = []
        dog=soup.select('td')
        for j in range(len(dog)):
            #한 종류의 강아지 정보를 저장하는 리스트
            temp.append(dog[j].text.strip())

        #전체리스트에 한 종류의 강아지 정보 추가
        dog_list.append(temp)
        time.sleep(1)
        driver.back()
    cntpage += 1
    
```

```python
#리스트길이가 13이면 7인덱스에 x추가
for i in range(len(dog_list)):
    if len(dog_list[i])==13:
        dog_list[i].insert(7,'x')
    
    else:
        dog_list[i]
dog_list
```

```python
dogdf=pd.DataFrame(dog_list)
```

```python
#페이지5에 있는 강아지 정보 크롤링
cntpage = 5
dog_list=[]
driver=wb.Chrome()

driver.get("https://terms.naver.com/list.naver?cid=42883&categoryId=44358&so=date.dsc&viewType=&categoryType=&page="+str(cntpage))
    
for i in range(12,16):
    btnDog=driver.find_element_by_css_selector("ul.content_list>li:nth-child("+str(i)+")>div.info_area>div.subject>strong.title>a:first-child")
    btnDog.click()
    time.sleep(1)
    soup=bs(driver.page_source,'lxml')

    #새로운 강아지 정보를 담아주기 위한 초기화
    temp = []
    dog=soup.select('td')
    for j in range(len(dog)):
        #한 종류의 강아지 정보를 저장하는 리스트
        temp.append(dog[j].text.strip())

    #전체리스트에 한 종류의 강아지 정보 추가
    dog_list.append(temp)
    time.sleep(1)
    driver.back()
```

#### 강아지 이름 크롤링하기


```python
#6~14페이지 강아지 이름 크롤링
cntpage = 6
dogName = []
driver=wb.Chrome()
for p in range(0,9):
    driver.get("https://terms.naver.com/list.naver?cid=42883&categoryId=44358&so=date.dsc&viewType=&categoryType=&page="+str(cntpage))
    
    for i in range(1,16):
        
        name=driver.find_element_by_css_selector("ul.content_list>li:nth-child("+str(i)+")>div.info_area>div.subject>strong.title>a:first-child").text   
        dogName.append(name)
    cntpage += 1
```
