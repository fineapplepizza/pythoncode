# 한솥도시락 이름,가격 정보 수집

```python
from selenium import webdriver as wb
from bs4 import BeautifulSoup as bs
import pandas as pd
import time
```

# 반복문의 진행상황 체크해주는 라이브러리

```python
from tqdm import tqdm_notebook as tq
```

```python
driver=wb.Chrome() #크롬창 그자체 모든화면전체정보를 드라이버가 가지고있다.비어있는 화면이 뜸
url = "https://www.hsd.co.kr/menu/menu_list"
driver.get(url)
time.sleep(0.3)
btn_more = driver.find_element_by_class_name("c_05")

try :
    for i in range(50) :
        btn_more.click() #클릭을 한 번 하고 나서 
        time.sleep(2)
except :
    print("종료되었습니다")
time.sleep(0.3)
soup=bs(driver.page_source,'lxml')
title=soup.select("h4.h.fz_03")
price=soup.select("div.item-price>strong")
time.sleep(0.3)
title_list=[]
price_list=[]
for i in tq(range(len(title))):
    title_list.append(title[i].text.strip())
    price_list.append(price[i].text.strip()+"원")
time.sleep(0.3)
data={"메뉴":title_list,"가격":price_list}
pd.DataFrame(data)
```

```python
soup=bs(driver.page_source,'lxml')
```

```python
#메뉴가져오기
title=soup.select("h4.h.fz_03")
```

```python
len(title)
```

```python
#가격가져오기
price=soup.select("div.item-price>strong")
```

```python
len(price)
```

```python
title_list=[]
price_list=[]
number_list=[]
for i in range(len(title)):
    title_list.append(title[i].text.strip())
    price_list.append(price[i].text.strip())
    number_list.append(i+1)
print(title_list)
print(price_list)
```

```python
#append 하고있는 리스트 변수명이 잘못되었습니다.감사합니다
```

```python
#데이터프레임제작
import pandas as pd
```

```python
data={"번호":number_list,"메뉴":title_list,"가격":price_list}
```

```python

pd.set_option('display.max_row',100)
```

```python
table=pd.DataFrame(data)
```
