# QUESTION 3

```python
from selenium import webdriver as wb
import pandas as pd
import time
from tqdm import tqdm_notebook as tq
```

```python
driver=wb.Chrome()
driver.get("http://corners.gmarket.co.kr/Bestsellers")
itemTitleList=[]
priceList=[]
rankList=[]
for i in tq(range(1,21)):
    item=driver.find_element_by_css_selector("p#no"+str(i)+"+div")
    item.click()
    time.sleep(1)
    itemTitle=driver.find_element_by_css_selector("h1.itemtit").text
    price=driver.find_element_by_css_selector("strong.price_real").text
    itemTitleList.append(itemTitle)
    priceList.append(price)
    rankList.append(i)
    time.sleep(1)
    driver.back()
    time.sleep(1)

info={"순위":rankList,"상품명":itemTitleList,"가격":priceList}
data=pd.DataFrame(info)
data.set_index("순위",inplace=True)
data



```

# Question 4

```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
user={"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.77 Safari/537.36"}
```

```python
res=req.get("https://www.melon.com/chart/index.htm",headers=user)
res.text
```

```python
soup=bs(res.text,'lxml')
```

```python
title=soup.select("div.ellipsis.rank01")
```

```python
len(title)
```

```python
title[0].text.strip()
```

```python
singer=soup.select("div.ellipsis.rank02")
```

```python
len(singer)
```

```python
title_list=[]
singer_list=[]
rank_list=[]
for i in range(len(singer)):
    title_list.append(title[i].text.strip())
    singer_list.append(singer[i].text.strip())
    rank_list.append(i+1)
```

```python
info={"순위":rank_list,"가수":singer_list,"제목":title_list}
```

```python
data=pd.DataFrame(info)
```

```python
data.set_index("순위",inplace=True)
```

```python
data
```

```python
data.to_csv("data.csv",encoding="")
```

# Question5

```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
import pandas as pd
from bs4 import BeautifulSoup as bs
import time
```

```python
driver=wb.Chrome()
driver.get("https://www.youtube.com/")
time.sleep(1)
search=driver.find_element_by_id("search")
time.sleep(1)
search.send_keys("지구를 지켜츄")
search.send_keys(Keys.ENTER)
time.sleep(1)
body=driver.find_element_by_tag_name("body")
for i in range(1,20):
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(1)
soup=bs(driver.page_source,'lxml')
title=soup.select("yt-formatted-string.style-scope.ytd-video-renderer")
count=soup.select("#metadata-line > span:nth-child(1)")
title_list=[]
count_list=[]
for i in range(len(count)):
    title_list.append(title[i].text.strip())
    count_list.append(count[i].text)
    time.sleep(0.3)
info={"제목":title_list,"조회수":count_list}
pd.DataFrame(info)
```

```python
data=pd.DataFrame(info)
```

```python
data.to_csv("youtube.csv",encoding="")
```
